```python
#missing values

# Step 1: rows where 'pizza_name_id' is null
null_rows = sales['pizza_name_id'].isnull()
```

```python
sales.loc[null_rows, 'pizza_name_id']
```

```python

# Step 2: Only fill the NaN values, keeping other values intact
sales.loc[null_rows, 'pizza_name_id'] = sales.groupby(['pizza_name', 'pizza_category', 'pizza_size'])['pizza_name_id'].transform(lambda x: x.ffill().bfill())[null_rows]
```
