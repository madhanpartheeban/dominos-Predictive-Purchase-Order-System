```python
import pandas as pd

# Load the Excel files
file1 = r"C:\Users\GS0864\Downloads\Pizza_Sale.xlsx"
file2 = r"C:\Users\GS0864\Downloads\Pizza_ingredients.xlsx"

# Convert the sheets to CSV
sheet1 = pd.read_excel(file1, sheet_name=None)  # Load all sheets as a dictionary
for sheet_name, df in sheet1.items():
    csv_file = f"{sheet_name}_Pizza_Sale.csv"
    df.to_csv(csv_file, index=False)
    print(f"Converted {sheet_name} to {csv_file}")

sheet2 = pd.read_excel(file2, sheet_name=None)  # Load all sheets as a dictionary
for sheet_name, df in sheet2.items():
    csv_file = f"{sheet_name}_Pizza_ingredients.csv"
    df.to_csv(csv_file, index=False)
    print(f"Converted {sheet_name} to {csv_file}")
```
