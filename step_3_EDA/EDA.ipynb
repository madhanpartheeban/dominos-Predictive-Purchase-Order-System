```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv(r"C:\Users\GS0864\OneDrive\Desktop\for_vs_code\DOMINOS\step_2_data_cleaning\Pizza_sales.csv", index_col=0)
ingredients = pd.read_csv(r"C:\Users\GS0864\OneDrive\Desktop\for_vs_code\DOMINOS\step_2_data_cleaning\ingredients_data.csv", index_col=0)
```

```python
sales['order_date'] = pd.to_datetime(sales['order_date'], format="mixed")
```

```python
sales.info()
```

```python
sales["day_of_week"] = sales['order_date'].dt.dayofweek
```

```python
sales.head(5)
```

```python
sales['month'] = sales['order_date'].dt.month
```

```python
sales.head(5)
```

```python
# EDA: Plot sales trends over time
sales.groupby('order_date')['quantity'].sum().plot()
plt.title('Total Sales Over Time')
plt.show()
```

```python
# Example of extracting features for time series modeling
sales['is_weekend'] = sales['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
sales.head(5)
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = sales)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = sales["quantity"])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = sales["unit_price"])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = sales["total_price"])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x="month",data=sales)

# plt.xticks(ticks=range(1, 13))

plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.title('Monthly Pizza Sales for 2015')

plt.show()
```

```python
df2 = sales.groupby(['month','pizza_category'])['quantity'].sum().reset_index()

sns.catplot(data=df2, x="pizza_category", y="quantity",hue="month", kind="bar",palette="dark", height=6)

plt.xlabel('Pizza Category')
plt.ylabel('Quantity Sold')
plt.title('Pizza category sales by month')

plt.show()
```

```python
df3 = sales.groupby(['order_date','pizza_category',"pizza_name"])['quantity'].sum().reset_index()

plt.figure(figsize=(20,6))
sns.barplot(x="pizza_name", y="quantity", data=df3, color="b")

plt.xlabel('Pizza Name')
plt.ylabel('Quantity Sold')
plt.title('Types of Pizza was sold')

plt.show()
```

```python
df4 = sales.groupby(['is_weekend','pizza_category'])['quantity'].sum().reset_index()

sns.barplot(x="is_weekend", y="quantity", data=df4, color="r")

plt.xlabel('Is Weekend')
plt.ylabel('Quantity Sold')
plt.title('Weekend Sales')

plt.show()
```
